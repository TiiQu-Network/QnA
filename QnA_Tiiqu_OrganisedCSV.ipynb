**Copy output q&a pair text, save in a text file, read text and Save q&a pairs on a csv file while placing them in their respectice columns (question and answer)**

```python
import csv

def parse_qa_pairs(lines):
    pairs = []
    current_q = ''
    current_a = ''
    for line in lines:
        line = line.strip()
        if line.startswith('Q1') or line.startswith('Q2'):
 
            if current_q:
                pairs.append((current_q, current_a))
            current_q = line
            current_a = ''
        elif line.startswith('A1') or line.startswith('A2'):
           
            current_a += line[2:] + '\n'
  
    if current_q:
        pairs.append((current_q, current_a))
    return pairs

with open('/content/drive/MyDrive/merged.txt', 'r', encoding='utf-8') as file:
    lines = file.readlines()
    pairs = parse_qa_pairs(lines)

with open('/content/drive/MyDrive/output21.csv', 'w', encoding='utf-8', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Question', 'Answer'])
    for Q, A in pairs:
        if Q.startswith('Q1') or Q.startswith('Q2'):
         
            writer.writerow([Q, A.strip()])
        elif q.startswith('A1') or q.startswith('A2'):
     
            writer.writerow([A.strip(), Q])
```
